## Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

## Import the csv file to dataframe

```python
test = pd.read_csv(r"F:\NeU 2nd Sem\ADM\Assignment 3\test (1).csv")
```

```python
## checking columns in dataframe

test.columns
```

```python
test.drop(["Unnamed: 1","A"] axis=1, inplace=True)
```

```python
test.head()
```

## Check correlations

```python
test.corr()
```

```python
# Generate a mask for the upper triangle

mask = np.zeros_like(test.corr(), dtype=np.bool)
#print(mask)
mask[np.triu_indices_from(mask)] = True
#print(mask)
plt.figure(figsize=(15,8))
sns.heatmap(test.corr(), cmap="jet", vmin=-1, vmax=1,center=0, annot=True)
```

## Plotting Sales vs Week

```python
## Week vs Base sales

plt.figure(figsize=(15,8))
plt.plot(test.week, test.base_sales)
plt.title("Week vs Base sales")
plt.xlabel("week")
plt.ylabel("base sales")
```

## Plotting all 4 different sales vs week



```python
plt.figure(figsize=(15,8))
plt.plot(test.week, test.base_sales, label="base sales")
plt.plot(test.week, test.tv_Sales_1, label="base sales + tv + adstock1")
plt.plot(test.week, test.tv_Sales_2, label="base sales + tv + adstock2")
plt.plot(test.week, test.tv_Radio_Sales_1, label="base sales + tv + radio + adstock1")
plt.plot(test.week, test.tv_Radio_Sales_2, label="base sales + tv + radio + adstock2")
plt.legend()
plt.xlabel("Weeks")
plt.ylabel("Sales")
plt.title("Sales vs Weeks")
```

## OLS Models

### Affect of TV spending, adstock model 1 on sales

```python
est = sm.OLS(test.tv_Sales_1, test[["base_sales", "temp", "tv_spend", "week", "adstock_TV_1"]])
est1 = est.fit()
print(est1.summary())
```

### Affect of tv spend, adstock model 2 on sales

```python
est2 = sm.OLS(test.tv_Sales_2, test[["base_sales", "temp", "tv_spend", "week", "tv_Sales_2_Adstock"]])
est3 = est2.fit()
print(est3.summary())
```

```python
## Effect of TV Spend on sales 

fig, ax = plt.subplots(figsize=(15,8))
ax.plot(test.week, test.tv_Sales_1, c="g", label="TV Sales 1")
ax.plot(test.week, test.tv_Sales_2, c="b", label="TV Sales 2")
ax.plot(test.week, test.base_sales, c="orange", label="Base sales")
ax.legend()
ax.tick_params(axis='y', labelcolor="green")
ax.set_ylabel("TV Sales 1")

ax2 = ax.twinx()
ax2.plot(test.week, test.tv_spend, color="red")
ax2.tick_params(axis='y', labelcolor="red")
ax2.set_ylabel("TV spend", c="r")
ax.set_xlabel("week")
fig.tight_layout()
ax.set_title("Effect of TV spend on sales")
```

## Effect of TV spend on Adstock models

```python
## plot adstock model 1 vs tv spend 

fig, ax = plt.subplots(figsize=(15,8))
ax.plot(test.week, test.adstock_TV_1, c="g",)
ax.tick_params(axis='y', labelcolor="green")
ax.set_ylabel("Adstock model 1", c="g")
ax.set_title("Effect of Adstock model 1 on TV spend")


ax2 = ax.twinx()
ax2.plot(test.week, test.tv_spend, color="red")
ax2.tick_params(axis='y', labelcolor="red")
ax2.set_ylabel("TV spend",c="r")

ax.set_xlabel("week")
fig.tight_layout()
```

```python
## plot adstock model 2 vs tv spend 

fig, ax = plt.subplots(figsize=(15,8))
ax.plot(test.week, test.adstock_TV_1, c="b", label="adstock model 1")
ax.plot(test.week, test.tv_Sales_2_Adstock, c="g", label="adstock model 2")
ax.tick_params(axis='y', labelcolor="green")
ax.set_ylabel("Adstock model 2", c="g")
ax.legend()

ax2 = ax.twinx()
ax2.plot(test.week, test.tv_spend, color="red")
ax2.tick_params(axis='y', labelcolor="red")
ax2.set_ylabel("TV spend", c="r")
ax.set_xlabel("week")
ax.set_title("Effect of adstock on TV Sales")
fig.tight_layout()
```

## Radio spending and sales

```python
est4 = sm.OLS(test.tv_Radio_Sales_1, test[["base_sales", "temp", "tv_spend", "radioSpend", "week", "tv_Radio_Sales_1_Adstock"]])
est5 = est4.fit()
print(est5.summary())
```

```python
est6 = sm.OLS(test.tv_Radio_Sales_2, test[["base_sales", "temp", "tv_spend", "radioSpend", "week", "tv_Radio_Sales_2_Adstock"]])
est7 = est6.fit()
print(est7.summary())
```
